```python
from rdflib import Graph, Literal, RDF, URIRef, Namespace
from rdflib.namespace import  FOAF, DC, RDF, RDFS, OWL, SKOS, NamespaceManager
from rdflib import URIRef, BNode, Literal
import csv
```

```python

g = Graph()

nm = NamespaceManager(g)

ingredient = URIRef("http://data.lirmm.fr/ontologies/food#Ingredient")
ingredient_ns = Namespace("https://kg.knowledgegraph.tech/food#Ingredient")
#food_ns = Namespace("https://kg.knowledgegraph.tech/food#")
#ingredient = food['Ingredient']

ingredients_list = csv.reader(open('ingredients_list.csv'))

i = 0
for row in ingredients_list:
    ingredient_uri = URIRef(ingredient_ns + str(i))
    g.add((ingredient_uri, RDF.type, ingredient))
    g.add((ingredient_uri, RDFS.label, Literal(row[0])))
    i += 1

g.serialize(destination = "kgc_food_ingredients.ttl", format = "turtle")
```
