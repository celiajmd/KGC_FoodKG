```python
import json
import os
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
# load ingredients from train 

ingredients_train = []    

with open("train.json") as f:
    ing_list = json.load(f)    
    for item in tqdm(ing_list, position=0):
        ingredients_train.append([ing for ing in item['ingredients']])
```

```python
# load ingredients from test

ingredients_test = []    

with open("test.json") as f:
    ing_list = json.load(f)    
    for item in tqdm(ing_list, position=0):
        ingredients_test.append([ing for ing in item['ingredients']])
```

```python
# all the ingredients with duplication
ingredients_all = ingredients_train + ingredients_test
```

```python
len(ingredients_all)
```

```python
# flatten ingredients_all

ingredients_flat = [item for sublist in ingredients_all for item in sublist]
```

```python
len(ingredients_flat)
```

```python
# de-duplicate: make the list to a set

ingredients_set = set(ingredients_flat)
```

```python
len(ingredients_set)
```

```python
ingredients_list = list(ingredients_set)
```

```python
len(ingredients_list)
```

```python
# write to a txt
with open('ingredients_set.txt','w') as d:
   d.write(str(ingredients_set))  
```

```python
# write to a csv
import csv
cw = csv.writer(open("ingredients_list.csv",'w'))
cw.writerow(ingredients_list)
```
