```python
# check reconciled ingredients 
ingredient_idx = {'almond': 0,
 'butter': 1,
 'hot pepper': 2,
 'cinnamon': 3,
 'coconut': 4,
 'cornstarch': 5,
 'cream': 6,
 'curry': 7,
 'egg': 8,
 'flour': 9,
 'garlic': 10,
 'gelatin': 11,
 'honey': 12,
 'molasses': 13,
 'nutmeg': 14,
 'raisin': 15,
 'tomato sauce': 16,
 'tomato': 17,
 'wheat': 18,
 'zucchini': 19}
```

```python
# create recipes
import csv
# Remember to check the header line
clean_recipe = csv.reader(open('recipes_OpenRefined.csv'))
```

```python
i = 0 #For ingredients
j = 0 #For recipes
recipes_idx = {} #For loop unique ingredient

for row in clean_recipe:
    recipe = row[2].replace("  ", " ")
    recipe = row[2].replace("   ", " ")
    recipe = row[2].replace("/", "or")
    recipe_lowercase = recipe.lower()
    if recipe_lowercase in recipes_idx or row[2] == 'Recipe Name_refined':
        continue
    #don't show empty images
    if row[4] == 'https://images.media-allrecipes.com/images/79591.png':
        continue
    #don't show invalid time
    if row[6] == 'X' or row[7] == 'X' or row[8] == 'X':
        continue
    if recipe != '':
        i += 1
        if i > 1000:
            break
        filename = 'recipe_'+ recipe + '.md'
        f = open(filename, "x")
        cur_ingredient = set()
        recipes_idx[recipe_lowercase] = i
        f.write("---\n")
        f.write("title: \"%s\"\n"%(recipe))
        f.write("description: \"This is a recipe\"\n")
        f.write("type: \"featured\"\n")
        f.write("image: %s\n"%(row[4]))
        f.write("categories: \n")
        f.write("  - \"Recipe\"\n")
        f.write("tags: \n")
        f.write("  - \"Recipe\"\n")
        f.write("---\n")
        f.write("\n")
        f.write("\n")
        f.write(">**If you don't know what are inside this recipe, now you know!**\n")
        f.write("\n")
        f.write("![](../images/Recipes-Banner.jpg)")
        f.write("\n")
        f.write("> **Prepare Time:** %s\n"%(row[6]))
        f.write("\n")
        f.write("\n")
        f.write("> **Cook Time:** %s\n"%(row[7]))
        f.write("\n")
        f.write("\n")
        f.write("> **Total Time:** %s\n"%(row[8]))
        f.write("\n")
        f.write("> **Look the ingredients of this recipe**\n")
        f.write("\n")
    # de-duplication
    if row[9] in cur_ingredient:
        continue
    if row[9] in ingredient_idx:  
        cur_ingredient.add(row[9])
        f.write("> [%s]({{< ref \"ingredient_%s.md\" >}} \"%s\")"%(row[9],row[9],row[9]))          
        f.write("\n")        
        f.write("\n")
    else: 
        cur_ingredient.add(row[9])
        f.write("> %s"%(row[9]))
        f.write("\n")
        f.write("\n")

f.close()
```

```python
# generate .md files for all the reconciled ingredients

i = 0 #For ingredients
ingredient_md = {}

for row in reconciled_ingredient:
    if row[1] == 'recipe':
        continue
    if row[3] not in ingredient_md:
        filename = 'ingredient_' + str(row[3]) + '.md'
        f = open(filename, "x")
        ingredient_md[row[3]] = i        
        f.write("---\n")
        f.write("title: \"%s\"\n"%(row[4]))
        f.write("description: \"This is an ingredient\"\n")
        f.write("type: \"post\"\n")
        f.write("image: \"%s\"\n"%(row[7]))
        f.write("categories: \n")
        f.write("  - \"Ingredient\"\n")
        f.write("tags: \n")
        f.write("  - \"Ingredient\"\n")
        f.write("---\n")
        f.write("\n")
#         f.write("Let's check out this ingredient\n")
        f.write("\n")
        f.write(">This ingredient has a Chinese name:%s\n"%(row[5]))
        f.write("\n")
        f.write("> It is a class of %s\n"%(row[9]))
        f.write("\n")
        f.write("> See what people are saying about %s on Quora:\n"%(row[4]))
        f.write("https://www.quora.com/topic/%s\n"%(row[8]))
        f.write("\n")
        f.write("![](../images/ingredient.jpg)")
        f.write("\n")
        f.write("Look what recipes use this ingredient:\n")
        f.write("\n")
        i += 1   
    recipe = row[2].replace("  ", " ")
    recipe = row[2].replace("   ", " ")
    recipe = row[2].replace("/", "or")    
    recipe_lowercase = recipe.lower()
    if recipe_lowercase in recipes_idx:  
        f.write("> [%s]({{< ref \"recipe_%s.md\" >}} \"%s\")"%(recipe,recipe,recipe))  
        f.write("\n")
        f.write("\n")
    else:        
        f.write("> %s"%(recipe))
        f.write("\n")
        f.write("\n")
#     f.write("> %s\n"%(row[2]))
#     f.write("\n")
f.close()
```
